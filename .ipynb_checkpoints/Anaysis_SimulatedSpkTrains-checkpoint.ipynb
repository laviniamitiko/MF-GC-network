```python
import matplotlib.pyplot as plt

def load_spikes_from_dat(filename):
    """
    Load spikes from .dat file formatted as: <cell_id> <spike_time>
    Returns a dictionary: {cell_id: [spike_times]}
    """
    spikes = {}
    with open(filename, 'r') as f:
        for line in f:
            if line.strip() == "":
                continue
            parts = line.strip().split()
            if len(parts) != 2:
                continue
            cell_id = int(parts[0])
            time = float(parts[1])
            if cell_id not in spikes:
                spikes[cell_id] = []
            spikes[cell_id].append(time)
    return spikes

def plot_raster(spikes, title, ax, ylabel="Neuron #", xlabel="Time (ms)"):
    """
    Plot raster of spike times.
    `spikes` is a dict {cell_id: [times]}
    """
    for neuron_id, times in spikes.items():
        ax.vlines(times, neuron_id + 0.5, neuron_id + 1.5)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_title(title)
```

```python
# Load spike data (update filenames if needed)
mf_spikes = load_spikes_from_dat("MF_spikes_states_2.dat")
grc_spikes = load_spikes_from_dat("GrC_spikes_states_2.dat")

# Create the figure
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot both rasters
plot_raster(mf_spikes, "Mossy Fiber Spikes", axs[0], ylabel="MF #")
plot_raster(grc_spikes, "Granule Cell Spikes", axs[1], ylabel="GrC #", xlabel="Time (s)")

plt.tight_layout()
#plt.xlim(10,11)
plt.show()
```

```python
import numpy as np

def randblock(x, block_size):
    """Shuffle x in non-overlapping blocks of block_size (blockwise permutation)."""
    x = np.array(x)
    n_blocks = len(x) // block_size
    blocks = np.array_split(x[:n_blocks * block_size], n_blocks)
    np.random.shuffle(blocks)
    return np.concatenate(blocks)

def bootstrap_cc(x1, x2, block_size, N_itr):
    """
    Bootstrap cross-correlation between two time series matrices (T x N).
    
    Parameters:
    x1, x2 : np.ndarray
        Input time series (T x N)
    block_size : int
        Size of shuffled blocks
    N_itr : int
        Number of bootstrap iterations

    Returns:
    cc0 : np.ndarray
        Original correlations (N,)
    zc : np.ndarray
        Z-scored correlations (N,)
    """
    x1 = np.asarray(x1)
    x2 = np.asarray(x2)
    
    # Compute original correlation column-wise
    cc0 = np.array([np.corrcoef(x1[:, i], x2[:, i])[0, 1] for i in range(x1.shape[1])])

    # Pre-allocate for shuffled correlations
    cc_sh = np.ones((N_itr, x1.shape[1]))
    
    # Blockwise shuffling
    ll1 = (x2.shape[0] // block_size) * block_size

    for itr in range(N_itr):
        x2_sh = x2.copy()
        for i in range(x2.shape[1]):
            x2_sh[:ll1, i] = randblock(x2[:ll1, i], block_size)
        cc_sh[itr, :] = [np.corrcoef(x1[:, i], x2_sh[:, i])[0, 1] for i in range(x1.shape[1])]

    # Compute z-scores
    zc = (cc0 - np.nanmean(cc_sh, axis=0)) / np.nanstd(cc_sh, axis=0)
    
    return cc0, zc
```

```python
import numpy as np
from scipy.ndimage import gaussian_filter1d

def spikes_to_rates(spikes, bin_width=0.01, t_start=None, t_end=None,
                    smooth_sigma=None, target_rate=None):
    """
    Convert spike times into firing rate time series, optionally downsampled to match behavior.

    Parameters:
    - spikes: dict {cell_id: [spike_times]}
    - bin_width: float, initial bin width in seconds (e.g., 0.01 for 10 ms)
    - t_start, t_end: float, time window (if None, inferred from spikes)
    - smooth_sigma: float, Gaussian smoothing in bins
    - target_rate: float or None, if set, downsample to this rate (e.g., 7.2 Hz)

    Returns:
    - rates_ds: np.ndarray [T', N] downsampled
    - time_bins_ds: np.ndarray [T'] corresponding time points
    """
    all_spike_times = np.concatenate(list(spikes.values()))
    if t_start is None:
        t_start = np.min(all_spike_times)
    if t_end is None:
        t_end = np.max(all_spike_times)

    time_bins = np.arange(t_start, t_end, bin_width)
    T = len(time_bins) - 1
    N = len(spikes)
    
    rates = np.zeros((T, N))
    neuron_ids = sorted(spikes.keys())
    
    for idx, neuron_id in enumerate(neuron_ids):
        spike_times = np.array(spikes[neuron_id])
        hist, _ = np.histogram(spike_times, bins=time_bins)
        rate = hist / bin_width
        if smooth_sigma is not None:
            rate = gaussian_filter1d(rate, sigma=smooth_sigma)
        rates[:, idx] = rate

    time_centers = time_bins[:-1] + bin_width / 2

    if target_rate is not None:
        # Downsample to match behavior
        target_dt = 1.0 / target_rate
        target_times = np.arange(t_start, t_end, target_dt)
        indices = np.searchsorted(time_centers, target_times)
        indices = indices[indices < len(time_centers)]
        rates = rates[indices]
        time_centers = time_centers[indices]

    return rates, time_centers
```

```python
def spikes_to_gcamp_signal(spikes, dt=0.0001, t_start=None, t_end=None, tau=0.5):
    """
    Convert spike trains into GCaMP6f-like calcium signals using exponential decay.

    Parameters:
    - spikes: dict {cell_id: [spike_times]}
    - dt: float, time resolution (s)
    - t_start, t_end: float, time limits (inferred if None)
    - tau: float, decay time constant in seconds (e.g., 0.5 for 500 ms)

    Returns:
    - signals: np.ndarray [T, N], GCaMP-like traces
    - time: np.ndarray [T]
    """
    import numpy as np
    from scipy.signal import convolve

    all_spike_times = np.concatenate(list(spikes.values()))
    if t_start is None:
        t_start = np.min(all_spike_times)
    if t_end is None:
        t_end = np.max(all_spike_times)
    
    time = np.arange(t_start, t_end, dt)
    T = len(time)
    N = len(spikes)

    # GCaMP6f exponential kernel (normalized)
    kernel_length = int(5 * tau / dt)
    kernel_time = np.arange(0, kernel_length * dt, dt)
    kernel = np.exp(-kernel_time / tau)
    kernel /= np.sum(kernel)

    # Convert spikes to GCaMP-like signal
    signals = np.zeros((T, N))
    neuron_ids = sorted(spikes.keys())
    
    for i, neuron_id in enumerate(neuron_ids):
        spike_train = np.zeros(T)
        spike_times = spikes[neuron_id]
        spike_indices = ((np.array(spike_times) - t_start) / dt).astype(int)
        spike_indices = spike_indices[(spike_indices >= 0) & (spike_indices < T)]
        spike_train[spike_indices] = 1.0
        signals[:, i] = convolve(spike_train, kernel, mode='full')[:T]

    return signals, time
```

```python
def downsample_to_behavior(signal, orig_dt, target_rate):
    from scipy.signal import resample

    orig_fs = 1 / orig_dt
    target_len = int(signal.shape[0] * target_rate / orig_fs)
    signal_ds = resample(signal, target_len, axis=0)
    time_ds = np.arange(target_len) / target_rate
    return signal_ds, time_ds
```

```python
import numpy as np
from scipy.io import loadmat

def load_behavior_trimmed(mat_file, acq_rate, minT, maxT):
    """
    Load and trim behavior data based on acquisition rate and desired time window.

    Returns:
    - behavior: [T, 1] array
    - time: [T] array in seconds
    """
    mat = loadmat(mat_file)
    if 'state_ds' not in mat:
        raise KeyError("'state_ds' not found in .mat file")

    state = mat['state_ds']
    if state.shape[0] == 1:
        state = state.T

    min_idx = int(minT * acq_rate)
    max_idx = int(maxT * acq_rate)

    behavior = state[min_idx:max_idx]
    time = np.arange(min_idx, max_idx) / acq_rate

    return behavior, time
```

```python
from scipy.signal import convolve

def spikes_to_gcamp_signal(spikes, dt=0.0001, tau=0.5):
    """
    Convert spike trains to GCaMP-like calcium signals using exponential decay.
    Spike times are assumed to be in seconds.
    """
    all_spike_times = np.concatenate(list(spikes.values()))
    t_start = 0
    t_end = np.max(all_spike_times)
    time = np.arange(t_start, t_end, dt)
    T = len(time)
    N = len(spikes)

    kernel_len = int(5 * tau / dt)
    kernel_time = np.arange(0, kernel_len * dt, dt)
    kernel = np.exp(-kernel_time / tau)
    kernel /= kernel.sum()

    signals = np.zeros((T, N))
    for i, (cell_id, times) in enumerate(sorted(spikes.items())):
        spike_train = np.zeros(T)
        idxs = (np.array(times) / dt).astype(int)
        idxs = idxs[idxs < T]
        spike_train[idxs] = 1
        signals[:, i] = convolve(spike_train, kernel, mode='full')[:T]

    return signals, time
```

```python
from scipy.signal import resample

def downsample_signal(signal, orig_dt, target_rate):
    """
    Resample [T, N] signal to match target rate (e.g. 7.2 Hz)
    """
    orig_fs = 1 / orig_dt
    T_new = int(signal.shape[0] * target_rate / orig_fs)
    signal_ds = resample(signal, T_new, axis=0)
    time_ds = np.arange(T_new) / target_rate
    return signal_ds, time_ds
```

```python
def randblock(x, block_size):
    x = np.array(x)
    n_blocks = len(x) // block_size
    blocks = np.array_split(x[:n_blocks * block_size], n_blocks)
    np.random.shuffle(blocks)
    return np.concatenate(blocks)

def bootstrap_cc(x1, x2, block_size, N_itr):
    x1 = np.asarray(x1)
    x2 = np.asarray(x2)
    
    cc0 = np.array([np.corrcoef(x1[:, i], x2[:, 0])[0, 1] for i in range(x1.shape[1])])
    cc_sh = np.zeros((N_itr, x1.shape[1]))

    ll1 = (x2.shape[0] // block_size) * block_size

    for itr in range(N_itr):
        x2_sh = x2.copy()
        x2_sh[:ll1, 0] = randblock(x2[:ll1, 0], block_size)
        for i in range(x1.shape[1]):
            cc_sh[itr, i] = np.corrcoef(x1[:, i], x2_sh[:, 0])[0, 1]

    zc = (cc0 - np.nanmean(cc_sh, axis=0)) / np.nanstd(cc_sh, axis=0)
    return cc0, zc
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_zscores(zc, label, th=2):
    """
    Plot histogram of z-scored correlations and return significant neuron indices.

    Parameters:
    - zc: array of z-scores (e.g. from bootstrap_cc)
    - label: label for plot and printed output

    Returns:
    - pos_idx: indices where z-score > 2
    - neg_idx: indices where z-score < -2
    """
    plt.figure(figsize=(8, 4))
    plt.hist(zc, bins=30, alpha=0.8, color='steelblue', edgecolor='black')
    plt.axvline(2, color='red', linestyle='--', label='Z = 2')
    plt.axvline(-2, color='red', linestyle='--')
    plt.title(f"Z-scored Correlation: {label}")
    plt.xlabel("Z-score")
    plt.ylabel("Count")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    pos_idx = np.where(zc > th)[0]
    neg_idx = np.where(zc < -th)[0]
    nsm_idx = np.where((zc < th) & (zc >-th))[0]

    print(f" {label}: {len(pos_idx)} neurons with Z > 2 ({len(pos_idx)/len(zc)*100:.2f}%)")
    print(f" {label}: {len(neg_idx)} neurons with Z < -2 ({len(neg_idx)/len(zc)*100:.2f}%)")

    return pos_idx, neg_idx, nsm_idx
```

```python
def classify_and_group_spikes(spikes, pm_idx, nm_idx, ns_idx):
    """
    Organize spike trains by modulation group.
    
    Parameters:
    - spikes: dict {neuron_id: [spike_times]}
    - pm_idx, nm_idx, ns_idx: arrays or lists of neuron indices

    Returns:
    - dict with keys "PM", "NM", "NS" each mapping to {neuron_index: spike_times}
    """
    spikes_by_group = {"PM": {}, "NM": {}, "NS": {}}
    
    for idx in pm_idx:
        idx = int(idx)  # <-- FIX
        if idx in spikes:
            spikes_by_group["PM"][idx] = spikes[idx]
    for idx in nm_idx:
        idx = int(idx)
        if idx in spikes:
            spikes_by_group["NM"][idx] = spikes[idx]
    for idx in ns_idx:
        idx = int(idx)
        if idx in spikes:
            spikes_by_group["NS"][idx] = spikes[idx]

    return spikes_by_group
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

def plot_spike_raster(spikes_by_group):
    color_map = {"PM": "red", "NM": "blue", "NS": "green"}

    fig, ax = plt.subplots(figsize=(15, 8))
    neuron_idx = 0

    for group in ["PM", "NM", "NS"]:
        if group not in spikes_by_group:
            continue
        for neuron_id in sorted(spikes_by_group[group]):
            times = spikes_by_group[group][neuron_id]
            ax.vlines(times, neuron_idx + 0.5, neuron_idx + 1.5, color=color_map[group])
            neuron_idx += 1

    ax.set_ylim(0.5, neuron_idx + 0.5)
    ax.set_xlabel("Time (ms)")
    ax.set_ylabel("Neuron index (sorted by modulation)")
    ax.set_title("Spike Raster by Modulation Type")
    ax.grid(True)

    # Legend
    legend_elements = [Line2D([0], [0], color=color_map[g], lw=2, label=g) for g in color_map]
    ax.legend(handles=legend_elements, loc='upper right')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_single_neuron_psth(spike_times, bin_width=0.1, t_min=0, t_max=None):
    """
    Computes PSTH for a single neuron.
    
    Parameters:
        spike_times: list or array of spike times (in seconds)
        bin_width: size of time bins (in seconds)
        t_min: start time for binning
        t_max: end time for binning (auto if None)
        
    Returns:
        bin_centers, firing_rates
    """
    spike_times = np.array(spike_times)
    if t_max is None:
        t_max = np.max(spike_times)
        
    bins = np.arange(t_min, t_max + bin_width, bin_width)
    counts, _ = np.histogram(spike_times, bins=bins)
    firing_rate = counts / bin_width  # convert to Hz

    bin_centers = bins[:-1] + bin_width / 2
    return bin_centers, firing_rate
```

```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt

# ------------------------------------------------
# Helper functions
# ------------------------------------------------

def load_behavior_trimmed(mat_file, acq_rate, minT, maxT):
    mat = loadmat(mat_file)
    if 'state_ds' not in mat:
        raise KeyError("'state_ds' not found in .mat file")

    state = mat['state_ds']
    if state.shape[0] == 1:
        state = state.T

    min_idx = int(minT * acq_rate)
    max_idx = int(maxT * acq_rate)

    behavior = state[min_idx:max_idx]
    time = np.arange(min_idx, max_idx) / acq_rate
    return behavior, time
```

```python
import numpy as np

def block_shuffle(x, block_size):
    """
    Shuffles time series data in blocks along the time axis.
    """
    T, N = x.shape
    assert T % block_size == 0, "Time dimension must be divisible by block_size"
    x = x.copy()
    num_blocks = T // block_size
    idx_blocks = np.random.permutation(num_blocks)
    idx = np.concatenate([np.arange(i * block_size, (i + 1) * block_size) for i in idx_blocks])
    return x[idx, :]

def bootstrap_cc(x1, x2, block_size, N_itr=1000):
    """
    Compute correlation coefficients between x1 and x2 (both shape T x N),
    and estimate a null distribution via block-wise shuffling.

    Parameters:
        x1: array (T x N)  neural activity (e.g. PSTH)
        x2: array (T x N)  behavioral variable broadcasted to same shape
        block_size: int  number of time bins per shuffle block
        N_itr: int  number of bootstrap iterations

    Returns:
        cc0: array (N,)  actual correlation coefficients
        zc: array (N,)  z-scored correlation coefficients
    """
    x1 = np.asarray(x1)
    x2 = np.asarray(x2)
    assert x1.shape == x2.shape, "Inputs must have the same shape"

    cc0 = np.array([np.corrcoef(x1[:, i], x2[:, i])[0, 1] for i in range(x1.shape[1])])
    cc_sh = np.ones((N_itr, x1.shape[1]))

    # Make sure length is divisible by block_size
    T = x1.shape[0]
    T_crop = (T // block_size) * block_size
    x1_crop = x1[:T_crop, :]
    x2_crop = x2[:T_crop, :]

    for k in range(N_itr):
        x1_sh = block_shuffle(x1_crop, block_size)
        for i in range(x1.shape[1]):
            cc_sh[k, i] = np.corrcoef(x1_sh[:, i], x2_crop[:, i])[0, 1]

    # Compute z-scored correlation
    zc = (cc0 - np.mean(cc_sh, axis=0)) / (np.std(cc_sh, axis=0) + 1e-10)
    return cc0, zc
```

```python
def plot_spike_raster(spikes_by_group):
    color_map = {"PM": "red", "NM": "blue", "NS": "green"}

    fig, ax = plt.subplots(figsize=(15, 8))
    neuron_idx = 0

    for group in ["PM", "NM", "NS"]:
        if group not in spikes_by_group:
            continue
        for neuron_id in sorted(spikes_by_group[group]):
            times = spikes_by_group[group][neuron_id]
            ax.vlines(times, neuron_idx + 0.5, neuron_idx + 1.5, color=color_map[group])
            neuron_idx += 1

    ax.set_ylim(0.5, neuron_idx + 0.5)
    ax.set_xlabel("Time (ms)")
    ax.set_ylabel("Neuron index (sorted by modulation)")
    ax.set_title("Spike Raster by Modulation Type")
    ax.grid(True)

    # Legend
    legend_elements = [Line2D([0], [0], color=color_map[g], lw=2, label=g) for g in color_map]
    ax.legend(handles=legend_elements, loc='upper right')

    plt.tight_layout()
    plt.show()
```

```python
# Replace NSM with NS everywhere

# Breakdown: for each GrC, count PM, NM, NS inputs
input_breakdown_by_grc = []
for grc_id in range(N_grc):
    gloms = np.where(conn_mat[:, grc_id])[0]
    mfs = [glom_mf_id[g] for g in gloms]
    mods = [mf_modulation_labels[mf] for mf in mfs]
    count = Counter(mods)
    input_breakdown_by_grc.append({
        'grc_class': grc_classes[grc_id],
        'PM': count['PM'],
        'NM': count['NM'],
        'NS': count['NS']
    })

# Aggregate counts by GrC class and by input composition
combo_counts = defaultdict(Counter)
for entry in input_breakdown_by_grc:
    combo = f"{entry['PM']}PM-{entry['NM']}NM-{entry['NS']}NS"
    combo_counts[entry['grc_class']][combo] += 1

# Generate all possible combinations of (PM, NM, NS) summing to d
possible_combos = []
for pm in range(d+1):
    for nm in range(d+1 - pm):
        ns = d - pm - nm
        possible_combos.append(f"{pm}PM-{nm}NM-{ns}NS")

# Helper to extract counts from combo label
def get_num(s):
    if s.endswith('PM') or s.endswith('NM'):
        return int(s[:-2])
    elif s.endswith('NS'):
        return int(s[:-2])
    else:
        raise ValueError(f"Unexpected suffix in {s}")

colors = {'PM': 'red', 'NM': 'blue', 'NS': 'green'}

# Plot histograms with same x-axis for each GrC class
for grc_class in combo_counts:
    counts = combo_counts[grc_class]

    pm_vals = []
    nm_vals = []
    ns_vals = []
    total_vals = []

    for label in possible_combos:
        count = counts[label] if label in counts else 0
        total_vals.append(count)

        pm_count, nm_count, ns_count = [get_num(x) for x in label.split('-')]
        pm_vals.append(count * pm_count)
        nm_vals.append(count * nm_count)
        ns_vals.append(count * ns_count)

    ind = np.arange(len(possible_combos))
    width = 0.6

    plt.figure(figsize=(12, 5))
    plt.bar(ind, ns_vals, color=colors['NS'], label='NS')
    plt.bar(ind, nm_vals, bottom=ns_vals, color=colors['NM'], label='NM')
    bottom_stack = np.array(ns_vals) + np.array(nm_vals)
    plt.bar(ind, pm_vals, bottom=bottom_stack, color=colors['PM'], label='PM')

    plt.xticks(ind, possible_combos, rotation=45, ha='right')
    plt.ylabel('Total number of inputs')
    plt.title(f"MF Input Composition for GrCs classified as {grc_class}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Rebuild mapping: input_combo  Counter of grc_class
from collections import defaultdict, Counter
import numpy as np
import matplotlib.pyplot as plt

input_combo_to_output = defaultdict(Counter)

for entry in input_breakdown_by_grc:
    combo = f"{entry['PM']}PM-{entry['NM']}NM-{entry['NS']}NS"
    input_combo_to_output[combo][entry['grc_class']] += 1

# Generate all possible combos as before
possible_combos = []
for pm in range(d+1):
    for nm in range(d+1 - pm):
        ns = d - pm - nm
        possible_combos.append(f"{pm}PM-{nm}NM-{ns}NS")

# Sort by canonical order if you prefer
possible_combos.sort()

# Plot
colors = {'PM': 'red', 'NM': 'blue', 'NS': 'green'}
grc_classes_order = ['NS', 'NM', 'PM']  # order in stack

ind = np.arange(len(possible_combos))
width = 0.6

# Initialize bottom stack
bottom_stack = np.zeros(len(possible_combos))

plt.figure(figsize=(14, 6))

# Plot stacked bars: NS first (bottom), then NM, then PM
for grc_class in grc_classes_order:
    class_vals = []
    for combo in possible_combos:
        count = input_combo_to_output[combo][grc_class] if combo in input_combo_to_output else 0
        class_vals.append(count)
    
    plt.bar(ind, class_vals, bottom=bottom_stack, color=colors[grc_class], label=grc_class)
    bottom_stack += np.array(class_vals)

plt.xticks(ind, possible_combos, rotation=45, ha='right')
plt.ylabel("Number of GrCs")
plt.xlabel("Input Combination")
plt.title("GrC Output Class by Mossy Fiber Input Combination")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def classify_and_group_spikes(spikes, grc_cells):
    """
    Organize spike trains by modulation group using grc_cells labels.

    Parameters:
    - spikes: dict {neuron_id: [spike_times]}
    - grc_cells: list or array where grc_cells[i] is the category ("PM", "NM", "NS", "Silent") of neuron i

    Returns:
    - dict with keys "PM", "NM", "NS" each mapping to {neuron_index: spike_times}
    """
    spikes_by_group = {"PM": {}, "NM": {}, "NS": {}}

    for idx, label in enumerate(grc_cells):
        if label in spikes_by_group and idx in spikes:
            spikes_by_group[label][idx] = spikes[idx]

    return spikes_by_group
```
