```python
import os
import pickle as pkl
import random
import numpy as np
import scipy.io as io
from datetime import datetime

import neuroml as nml
from pyneuroml import pynml
from pyneuroml.lems.LEMSSimulation import LEMSSimulation
import lems.api as lems

runID = '1'  # Use a string to match file naming convention
correlationRadius = 0
NADT = 0

duration = 180  # ms
dt = 0.05       # ms
minimumISI = 2  # ms
ONRate = 50     # Hz
OFFRate = 0     # Hz

base_dir = 'tempdata'

with open('../params_file.pkl', 'rb') as file:
    params = pkl.load(file)

N_syn = params['N_syn'][int(runID) - 1]
f_mf = params['f_mf'][int(runID) - 1]
run_num = params['run_num'][int(runID) - 1]

print(f"N_syn: {N_syn}, f_mf: {f_mf}, run_num: {run_num}")

conn_file = f'network_structures/GCLconnectivity_{N_syn}.pkl'
with open(conn_file, 'rb') as file:
    conn_data = pkl.load(file)

conn_mat = conn_data['conn_mat']
N_mf, N_grc = conn_mat.shape

assert np.all(conn_mat.sum(axis=0) == N_syn), 'Connectivity matrix is invalid.'

print(f"N_mf: {N_mf}, N_grc: {N_grc}")

if correlationRadius == 0:
    N_mf_ON = int(N_mf * f_mf)
    mf_indices_ON = sorted(random.sample(range(N_mf), N_mf_ON))
else:
    f_mf_range = np.linspace(.05, .95, 19)
    f_mf_ix = np.where(f_mf_range == f_mf)[0][0]
    mat_data = io.loadmat(f'../../input_statistics/mf_patterns_r{correlationRadius}.mat')
    R = mat_data['Rs'][:, :, f_mf_ix]
    g = mat_data['gs'][f_mf_ix]
    t = np.dot(R.T, np.random.randn(N_mf))
    S = (t > -g * np.ones(N_mf))
    mf_indices_ON = np.where(S)[0].tolist()

N_mf_ON = len(mf_indices_ON)
mf_indices_OFF = sorted(set(range(N_mf)) - set(mf_indices_ON))
N_mf_OFF = len(mf_indices_OFF)

print(f"ON MFs: {N_mf_ON}, OFF MFs: {N_mf_OFF}")

spike_gen_file = '../grc_lemsDefinitions/spikeGenerators.xml'
iaf_file = f"../grc_lemsDefinitions/IaF_GrC.nml" if NADT == 0 else f"../grc_lemsDefinitions/IaF_GrC_{f_mf:.2f}.nml"
ampa_file = f"../grc_lemsDefinitions/RothmanMFToGrCAMPA_{N_syn}.xml"
nmda_file = f"../grc_lemsDefinitions/RothmanMFToGrCNMDA_{N_syn}.xml"

spike_doc = pynml.read_lems_file(spike_gen_file)
iaf_doc = pynml.read_neuroml2_file(iaf_file)
ampa_doc = pynml.read_lems_file(ampa_file)
nmda_doc = pynml.read_lems_file(nmda_file)

spike_type = spike_doc.component_types['spikeGeneratorRefPoisson']
lems_doc = lems.Model()

spike_on = lems.Component("mossySpikerON", spike_type.name)
spike_on.set_parameter("minimumISI", f"{minimumISI} ms")
spike_on.set_parameter("averageRate", f"{ONRate} Hz")
lems_doc.add(spike_on)

spike_off = lems.Component("mossySpikerOFF", spike_type.name)
spike_off.set_parameter("minimumISI", f"{minimumISI} ms")
spike_off.set_parameter("averageRate", f"{OFFRate} Hz")
lems_doc.add(spike_off)

grc_cell = iaf_doc.iaf_ref_cells[0]
net = nml.Network(id="network")
net_doc = nml.NeuroMLDocument(id="network")
net_doc.networks.append(net)

GrCPop = nml.Population(id="GrCPop", component=grc_cell.id, size=N_grc)
mossyON = nml.Population(id="mossySpikerONPop", component=spike_on.id, size=N_mf_ON)
mossyOFF = nml.Population(id="mossySpikerOFFPop", component=spike_off.id, size=N_mf_OFF)
net.populations.extend([GrCPop, mossyON, mossyOFF])

syns = [ampa_doc.components['RothmanMFToGrCAMPA'].id,
        nmda_doc.components['RothmanMFToGrCNMDA'].id]

# Connect ON MFs
for ix_on, mf_ix in enumerate(mf_indices_ON):
    for grc_ix in np.where(conn_mat[mf_ix, :] == 1)[0]:
        for syn in syns:
            net.synaptic_connections.append(nml.SynapticConnection(
                from_=f"{mossyON.id}[{ix_on}]",
                to=f"GrCPop[{grc_ix}]", synapse=syn))

# Connect OFF MFs
for ix_off, mf_ix in enumerate(mf_indices_OFF):
    for grc_ix in np.where(conn_mat[mf_ix, :] == 1)[0]:
        for syn in syns:
            net.synaptic_connections.append(nml.SynapticConnection(
                from_=f"{mossyOFF.id}[{ix_off}]",
                to=f"GrCPop[{grc_ix}]", synapse=syn))



os.makedirs(base_dir, exist_ok=True)
net_file = os.path.join(base_dir, f'generated_network_{runID}.net.nml')
lems_inst_file = os.path.join(base_dir, f'instances_{runID}.xml')

net_doc.includes.append(nml.IncludeType(href=os.path.basename(spike_gen_file)))
net_doc.includes.append(nml.IncludeType(href=os.path.basename(ampa_file)))
net_doc.includes.append(nml.IncludeType(href=os.path.basename(nmda_file)))

pynml.write_neuroml2_file(net_doc, net_file)
pynml.write_lems_file(lems_doc, lems_inst_file, validate=False)

# Create and run simulation
sim_id = f"sim_{runID}"
ls = LEMSSimulation(sim_id, duration, dt)
ls.assign_simulation_target("network")


ls.include_neuroml2_file(iaf_file)
ls.include_lems_file(spike_gen_file, include_included=False)
ls.include_lems_file(lems_inst_file)
ls.include_lems_file(ampa_file, include_included=False)
ls.include_lems_file(nmda_file, include_included=False)
ls.include_neuroml2_file(net_file)

lems_file = ls.save_to_file()
print("Running simulation...")
results = pynml.run_lems_with_jneuroml(lems_file, nogui=True, load_saved_data=False, plot=False)
```
