```python
import pickle as pkl
import numpy as np
import random
from neuroml import (
    NeuroMLDocument, Network, Population, IncludeType,
    SpikeGeneratorRefPoisson, SynapticConnection
)
from neuroml.utils import component_factory
from pyneuroml.pynml import write_neuroml2_file, run_lems_with_jneuroml
from pyneuroml.lems.LEMSSimulation import LEMSSimulation

# -------------------------------
# 1. Load connectivity + define MF activity
# -------------------------------
with open("../network_structures/GCLconnectivity_4.pkl", 'rb') as f:
    conn_mat = pkl.load(f)['conn_mat']

N_mf, N_grc = conn_mat.shape
N_syn = 4
assert np.all(conn_mat.sum(axis=0) == N_syn), "Invalid GrC inputs"

f_mf = 0.8
N_mf_ON = int(N_mf * f_mf)
mf_indices_ON = sorted(random.sample(range(N_mf), N_mf_ON))
mf_indices_OFF = sorted(set(range(N_mf)) - set(mf_indices_ON))

# -------------------------------
# 2. Choose GrC model based on NADT flag
# -------------------------------
use_nadt = True  # Enable NADT (activity-dependent Golgi effect)

# Select the appropriate GrC model file
if use_nadt:
    grc_cell_file = f"IaF_GrC_{f_mf:.2f}.nml"  # e.g., IaF_GrC_0.30.nml
else:
    grc_cell_file = "IaF_GrC.nml"

grc_cell_id = "IaF_GrC"  # Keep the internal ID consistent

# -------------------------------
# 3. Create NeuroML document and add populations
# -------------------------------
doc = NeuroMLDocument(id="MF_GrC_NetworkDoc")
net = Network(id="MF_GrC_Network")
doc.networks.append(net)

# Include the GrC model with or without NADT
doc.includes.append(IncludeType(href=grc_cell_file))

# Add GrC population
net.populations.append(Population(id="GrCPop", component=grc_cell_id, size=N_grc))

# Add Poisson spike generators
poisson_on = component_factory(SpikeGeneratorRefPoisson, id="PoissonON", average_rate="50 Hz", minimum_isi="2ms")
poisson_off = component_factory(SpikeGeneratorRefPoisson, id="PoissonOFF", average_rate="0 Hz", minimum_isi="2ms")
doc.spike_generator_ref_poissons.extend([poisson_on, poisson_off])

# Add MF populations
net.populations.append(Population(id="MossyONPop", component="PoissonON", size=len(mf_indices_ON)))
net.populations.append(Population(id="MossyOFFPop", component="PoissonOFF", size=len(mf_indices_OFF)))

# -------------------------------
# 4. Synaptic connections
# -------------------------------
for i, mf_idx in enumerate(mf_indices_ON):
    for grc_ix in np.where(conn_mat[mf_idx] == 1)[0]:
        for syn in ["RothmanMFToGrCAMPA", "RothmanMFToGrCNMDA"]:
            net.synaptic_connections.append(
                SynapticConnection(from_="MossyONPop[%d]" % i, to="GrCPop[%d]" % grc_ix, synapse=syn)
            )

for i, mf_idx in enumerate(mf_indices_OFF):
    for grc_ix in np.where(conn_mat[mf_idx] == 1)[0]:
        for syn in ["RothmanMFToGrCAMPA", "RothmanMFToGrCNMDA"]:
            net.synaptic_connections.append(
                SynapticConnection(from_="MossyOFFPop[%d]" % i, to="GrCPop[%d]" % grc_ix, synapse=syn)
            )

write_neuroml2_file(doc, "grc_full_network.net.nml")

# -------------------------------
# 5. Create and run LEMS simulation
# -------------------------------
sim = LEMSSimulation("Sim_MF_GrC", duration=180, dt=0.05)
sim.assign_simulation_target("MF_GrC_Network")
sim.include_neuroml2_file("grc_full_network.net.nml")
sim.include_neuroml2_file(grc_cell_file)
sim.include_lems_file("../grc_lemsDefinitions/RothmanMFToGrCAMPA_4.xml", include_included=False)
sim.include_lems_file("../grc_lemsDefinitions/RothmanMFToGrCNMDA_4.xml", include_included=False)

# Output files
sim.create_event_output_file("MFspikes_file", "MF_spikes.dat")
for i in range(len(mf_indices_ON)):
    sim.add_selection_to_event_output_file("MFspikes_file", i, "MossyONPop[%d]" % i, "spike")
for i in range(len(mf_indices_OFF)):
    sim.add_selection_to_event_output_file("MFspikes_file", len(mf_indices_ON) + i, "MossyOFFPop[%d]" % i, "spike")

sim.create_event_output_file("GrCspikes_file", "GrC_spikes.dat")
for i in range(N_grc):
    sim.add_selection_to_event_output_file("GrCspikes_file", i, "GrCPop[%d]" % i, "spike")

lems_file = sim.save_to_file()
results = run_lems_with_jneuroml(lems_file, max_memory="8G", nogui=True, plot=False)
```

```python
import matplotlib.pyplot as plt

def load_spikes_from_dat(filename):
    """
    Load spikes from .dat file formatted as: <cell_id> <spike_time>
    Returns a dictionary: {cell_id: [spike_times]}
    """
    spikes = {}
    with open(filename, 'r') as f:
        for line in f:
            if line.strip() == "":
                continue
            parts = line.strip().split()
            if len(parts) != 2:
                continue
            cell_id = int(parts[0])
            time = float(parts[1])
            if cell_id not in spikes:
                spikes[cell_id] = []
            spikes[cell_id].append(time)
    return spikes

def plot_raster(spikes, title, ax, ylabel="Neuron #", xlabel="Time (ms)"):
    """
    Plot raster of spike times.
    `spikes` is a dict {cell_id: [times]}
    """
    for neuron_id, times in spikes.items():
        ax.vlines(times, neuron_id + 0.5, neuron_id + 1.5)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_title(title)

# Load spike data (update filenames if needed)
mf_spikes = load_spikes_from_dat("MF_spikes.dat")
grc_spikes = load_spikes_from_dat("GrC_spikes.dat")

# Create the figure
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot both rasters
plot_raster(mf_spikes, "Mossy Fiber Spikes", axs[0], ylabel="MF #")
plot_raster(grc_spikes, "Granule Cell Spikes", axs[1], ylabel="GrC #", xlabel="Time (ms)")

plt.tight_layout()
plt.show()
```
