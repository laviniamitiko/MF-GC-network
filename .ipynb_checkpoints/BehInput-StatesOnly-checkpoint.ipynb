```python
import numpy as np
import neuroml as nml
from neuroml import NeuroMLDocument, SpikeArray, Spike
from pyneuroml import pynml
import pickle
from scipy.io import loadmat

def generate_rates_from_behavior(
    beh_file, minT, maxT, n_mf,
    acquisition_rate,              
    scale=50, offset=7, minrate=2,
    sparsity=1, weights=None, seed=None
):
    """
    Generate time-varying firing rates modulated by behavioral variables
    from a .mat file containing 'L_state' (1 x T or T x 1).
    
    Time is assumed to be sample indices, and converted using the acquisition rate.
    """
    if seed is not None:
        np.random.seed(seed)

    # Load behavior
    mat = loadmat(beh_file)
    if 'state_ds' not in mat:
        raise ValueError("Variable 'L_state' not found in the .mat file.")

    beh = mat['state_ds']

    # Ensure shape is (T, D) where T = timepoints, D = features
    if beh.shape[0] == 1:
        beh = beh.T  # from (1, T) to (T, 1)
    elif beh.shape[1] == 1:
        pass  # already (T, 1)
    else:
        raise ValueError(f"'L_state' shape is unexpected: {beh.shape}. Expected 1D signal.")

    # Convert time to indices
    min_idx = int(minT * acquisition_rate)
    max_idx = int(maxT * acquisition_rate)

    # Slice time range
    beh = beh[min_idx:max_idx, :]  # shape: (T, 1)
    T, D = beh.shape

    time = np.arange(min_idx, max_idx) / acquisition_rate  # shape: (T,)

    # Generate weights
    if weights is None:
        weights = np.random.rand(n_mf, D)
        mask = np.random.rand(n_mf, D) < sparsity
        weights[mask] = 0
    else:
        assert weights.shape == (n_mf, D), "Provided weights must match (n_mf, D)"

    # Compute rates
    rates = beh @ weights.T * scale + offset
    rates[rates < minrate] = minrate

    return time, rates, weights

def sample_poisson_spikes(rates, time, dt=0.05):
    """
    Sample inhomogeneous Poisson spike trains given rate profiles.
    """
    N = rates.shape[1]
    spike_trains = []

    for i in range(N):
        p = rates[:, i] * dt  # probability per time bin
        spikes = np.random.rand(len(p)) < p
        spike_times = time[spikes]
        spike_trains.append(spike_times)

    return spike_trains

#def getrate(rate, x, rate_dt):
#    """
#    Return rate values at times x using zero-order hold (sample-and-hold).
#
#    Parameters:
#    - rate: array of sampled rates (sampled at regular intervals `rate_dt`)
#    - x: array of time points (in seconds) where rate should be evaluated
#    - rate_dt: sampling interval for `rate` in seconds
#
#    Returns:
#    - r: array of rate values at each time in `x`
#    """
#    max_idx = len(rate) - 1
#    indices = np.floor(np.array(x) / rate_dt).astype(int)
#    indices = np.clip(indices, 0, max_idx)
#    return rate[indices]

def getrate(rate, x, rate_dt):
    """
    Return rate values at arbitrary time points x using sample-and-hold.

    Parameters:
    - rate: 1D array of firing rates (sampled at uniform rate_dt)
    - x: times at which to evaluate the rate
    - rate_dt: time resolution of `rate` (in seconds)

    Returns:
    - Array of rate values at each x
    """
    rate = np.asarray(rate).flatten()
    indices = np.floor(np.array(x) / rate_dt).astype(int)
    indices = np.clip(indices, 0, len(rate) - 1)
    return rate[indices]


def create_spike_array(rate, maxT, spike_dt, rate_dt, name):
    """
    Generate a spike array using inhomogeneous Poisson sampling.

    Parameters:
    - rate: array of firing rates (sampled at intervals of `rate_dt`)
    - maxT: duration of simulation (in seconds)
    - spike_dt: time step for evaluating spikes
    - rate_dt: resolution of `rate` array
    - name: ID of the spike array

    Returns:
    - neuroml.SpikeArray object
    """
    x = np.arange(0, maxT, spike_dt)
    m = getrate(rate, x, rate_dt)

    m2 = np.max(m)
    n_spikes = int(np.ceil(1.5 * maxT * m2))
    u = np.random.random((n_spikes, 1))
    y = np.cumsum(-np.log(u) / m2)
    y = y[y <= maxT]

    # thinning step
    p = np.random.rand(len(y))
    acceptance_probs = (getrate(rate, y, rate_dt) / m2).flatten()
    y_kept = y[p < acceptance_probs]

    # Create NeuroML spike array
    spk_arr = nml.SpikeArray(id=name)
    for i, t in enumerate(y_kept):
        spk_arr.spikes.append(nml.Spike(id=i, time=f"{1e3 * t:.3f} ms"))  # convert to ms

    return spk_arr


def write_combined_spikearray_nml(spike_groups, modulation_labels, filename="MF_BehMod.nml"):
    """
    Save spike trains to a NeuroML file with SpikeArrays grouped by modulation.
    """
    doc = NeuroMLDocument(id="MF_BehMod")

    for group, label in zip(spike_groups, modulation_labels):
        for i, spikes in enumerate(group):
            sa = SpikeArray(id=f"MF_{label}_{i}")
            for j, t in enumerate(spikes):
                sa.spikes.append(Spike(id=j, time=f"{t*1000:.3f} ms"))  # convert to ms
            doc.spike_arrays.append(sa)

    pynml.write_neuroml2_file(doc, filename)
    print(f"✅ Saved spike trains to: {filename}")
```

```python
import numpy as np
import pickle
from neuroml import NeuroMLDocument
from pyneuroml import pynml

# === Load behavior-modulated rate generator and Poisson spike functions ===
# Assumes you have:
# - generate_rates_from_behavior()
# - getrate()
# - create_spike_array(rate, maxT, spike_dt, rate_dt, name)

# === 1. Load connectivity to get number of MFs ===
with open("../network_structures/GCLconnectivity_4.pkl", 'rb') as f:
    conn_mat = pickle.load(f, encoding='latin1')['conn_mat']
N_mf = conn_mat.shape[0]

# === 2. Define group sizes ===
n_pm = int(N_mf * 0.4)
n_nm = int(N_mf * 0.3)
n_ns = N_mf - n_pm - n_nm

# === 3. Rate and spike generation settings ===
minT, maxT = 100, 140  # seconds
acq_rate = 7.2       # Hz
rate_dt = 1 / acq_rate  # seconds
spike_dt = 0.001     # seconds (1 ms resolution for spikes)
duration = maxT - minT

modulation_labels = ["PM", "NM", "NS"]
spike_arrays_all = []

# === 4. Generate spikes per group ===

# PM: positively modulated
time_pm, rates_pm, _ = generate_rates_from_behavior(
    beh_file='191018_13_39_41_dfTrace.mat',
    minT=minT, maxT=maxT, n_mf=n_pm,
    acquisition_rate=acq_rate,
    scale=, offset=7, minrate=2,
    sparsity=0, seed=1
)
for i in range(n_pm):
    sa = create_spike_array(rates_pm[:, i], maxT=duration, spike_dt=spike_dt, rate_dt=rate_dt, name=f"MF_PM_{i}")
    spike_arrays_all.append(sa)

# NM: negatively modulated
time_nm, rates_nm, _ = generate_rates_from_behavior(
    beh_file='191018_13_39_41_dfTrace.mat',
    minT=minT, maxT=maxT, n_mf=n_nm,
    acquisition_rate=acq_rate,
    scale=-25, offset=30, minrate=2,
    sparsity=0, seed=2
)
for i in range(n_nm):
    sa = create_spike_array(rates_nm[:, i], maxT=duration, spike_dt=spike_dt, rate_dt=rate_dt, name=f"MF_NM_{i}")
    spike_arrays_all.append(sa)

# NS: non-significantly modulated
time_ns, rates_ns, _ = generate_rates_from_behavior(
    beh_file='191018_13_39_41_dfTrace.mat',
    minT=minT, maxT=maxT, n_mf=n_ns,
    acquisition_rate=acq_rate,
    scale=0, offset=7, minrate=2,
    sparsity=0, seed=3
)
for i in range(n_ns):
    sa = create_spike_array(rates_ns[:, i], maxT=duration, spike_dt=spike_dt, rate_dt=rate_dt, name=f"MF_NS_{i}")
    spike_arrays_all.append(sa)

# === 5. Save to NeuroML ===
doc = NeuroMLDocument(id="MF_StatesOnly")
for sa in spike_arrays_all:
    doc.spike_arrays.append(sa)

pynml.write_neuroml2_file(doc, "MF_StatesOnly.nml")
print("✅ Saved behavior-modulated MF spike arrays to 'MF_StatesOnly.nml'")
```

```python
import matplotlib.pyplot as plt

def plot_example_rates(time_pm, rates_pm, time_nm, rates_nm, time_ns, rates_ns, n_show=5):
    """
    Plot example firing rate traces for PM, NM, and NS fibers.
    
    Parameters:
    - time_*(T,), rates_*(T x N): time vector and rate matrix for each modulation class
    - n_show: number of fibers to plot from each group
    """
    plt.figure(figsize=(15, 8))

    # Plot positively modulated (PM)
    for i in range(min(n_show, rates_pm.shape[1])):
        plt.plot(time_pm, rates_pm[:, i], label=f'PM {i}', alpha=0.8)

    # Plot negatively modulated (NM)
    for i in range(min(n_show, rates_nm.shape[1])):
        plt.plot(time_nm, rates_nm[:, i], label=f'NM {i}', alpha=0.8)

    # Plot non-significantly modulated (NS)
    for i in range(min(n_show, rates_ns.shape[1])):
        plt.plot(time_ns, rates_ns[:, i], label=f'NS {i}', alpha=0.8)

    plt.xlabel("Time (s)")
    plt.ylabel("Firing rate (Hz)")
    plt.title("Example Mossy Fiber Firing Rates")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_example_rates(time_pm, rates_pm, time_nm, rates_nm, time_ns, rates_ns)
```

```python
from neuroml.loaders import read_neuroml2_file

def load_spikes_from_nml(nml_file):
    """
    Load spike times from a NeuroML (.nml) file containing SpikeArray entries.

    Returns:
    - spikes_by_group: dict like {'PM': {0: [times]}, 'NM': {...}, 'NS': {...}}
    """
    doc = read_neuroml2_file(nml_file)

    spikes_by_group = {}

    for sa in doc.spike_arrays:
        sa_id = sa.id  # e.g., "MF_PM_21"
        # Extract modulation type and neuron index
        parts = sa_id.split('_')
        if len(parts) < 3:
            continue  # skip malformed IDs
        group = parts[1]  # PM, NM, NS
        idx = int(parts[2])  # neuron index

        if group not in spikes_by_group:
            spikes_by_group[group] = {}

        # Extract times and convert from "123.45 ms" -> float
        times = []
        for spike in sa.spikes:
            t = spike.time
            if isinstance(t, str) and t.endswith(" ms"):
                t = float(t.replace(" ms", ""))
            else:
                t = float(t)
            times.append(t)

        spikes_by_group[group][idx] = times

    return spikes_by_group
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

def plot_spike_raster(spikes_by_group):
    color_map = {"PM": "red", "NM": "blue", "NS": "green"}

    fig, ax = plt.subplots(figsize=(15, 8))
    neuron_idx = 0

    for group in ["PM", "NM", "NS"]:
        if group not in spikes_by_group:
            continue
        for i in sorted(spikes_by_group[group]):
            times = spikes_by_group[group][i]
            ax.vlines(times, neuron_idx + 0.5, neuron_idx + 1.5, color=color_map[group])
            neuron_idx += 1

    ax.set_ylim(0.5, neuron_idx + 0.5)
    ax.set_xlabel("Time (ms)")
    ax.set_ylabel("Neuron index")
    ax.set_title("Spike Raster from NeuroML")
    ax.grid(True)

    # Add legend
    legend_elements = [Line2D([0], [0], color=color_map[k], lw=2, label=k) for k in color_map]
    ax.legend(handles=legend_elements, loc='upper right')
    plt.tight_layout()
    #plt.xlim(250000, 350000)
    plt.show()
```

```python
spikes = load_spikes_from_nml("MF_StatesOnly.nml")
```

```python
plot_spike_raster(spikes)
```
