```python
import numpy as np
import neuroml as nml
from neuroml import NeuroMLDocument, SpikeArray, Spike
from pyneuroml import pynml
import pickle


def generate_rates_from_behavior(
    beh_file, minT, maxT, n_mf,
    scale=50, offset=7, minrate=2,
    sparsity=0.5, weights=None, seed=None
):
    """
    Generate time-varying firing rates modulated by behavioral variables.
    """
    if seed is not None:
        np.random.seed(seed)

    # Load behavior: columns = [time(ms), b1, b2, ..., bD]
    beh = np.genfromtxt(beh_file, delimiter=',')
    beh = beh[(beh[:, 0] >= minT * 1e3) & (beh[:, 0] < maxT * 1e3)]
    time = beh[:, 0] / 1000.0  # convert to seconds
    b_t = beh[:, 1:]           # T x D

    T, D = b_t.shape

    # Generate sparse weights
    if weights is None:
        weights = np.random.rand(n_mf, D)
        mask = np.random.rand(n_mf, D) < sparsity
        weights[mask] = 0
    else:
        assert weights.shape == (n_mf, D)

    # Compute rates: r_k(t) = scale * (w_k . b(t)) + offset
    rates = b_t @ weights.T * scale + offset
    rates[rates < minrate] = minrate

    return time, rates, weights


def sample_poisson_spikes(rates, time, dt=0.05):
    """
    Sample inhomogeneous Poisson spike trains given rate profiles.
    """
    N = rates.shape[1]
    spike_trains = []

    for i in range(N):
        p = rates[:, i] * dt  # probability per time bin
        spikes = np.random.rand(len(p)) < p
        spike_times = time[spikes]
        spike_trains.append(spike_times)

    return spike_trains

def getrate(rate, x, rate_dt):
    """
    Return rate values at times x using zero-order hold (sample-and-hold).

    Parameters:
    - rate: array of sampled rates (sampled at regular intervals `rate_dt`)
    - x: array of time points (in seconds) where rate should be evaluated
    - rate_dt: sampling interval for `rate` in seconds

    Returns:
    - r: array of rate values at each time in `x`
    """
    max_idx = len(rate) - 1
    indices = np.floor(np.array(x) / rate_dt).astype(int)
    indices = np.clip(indices, 0, max_idx)
    return rate[indices]


def create_spike_array( rate, maxT, dt, name ):
    '''
    Create inhomogeneous Poisson process
    Based on: https://hpaulkeeler.com/simulating-an-inhomogeneous-poisson-point-process/

    Inputs:
    - rate (array of rate sampled at dt)
    - maxT [FLOAT] = maximum spiketime (sec)
    - dt [FLOAT] = binning used to generate points (sec) (~0.1s is fine, smaller for faster varying rates)
    - name [STR] = Name of spike array
    '''

    x = np.arange( 0, maxT, dt )
    m = getrate( rate, x, dt ) #[0:len(x)]

    m2 = np.max(m)  # or just max(m)
    # generate uniform random variable
    u = np.random.random( size=(int(np.ceil(1.5*maxT*m2)),1) )
    y = np.cumsum( -1/m2 * np.log(u) ) # transform to homogeneous Poisson
    y = y[y<=maxT]

    # filter some points with probability = rate(t)/maxrate
    prob = np.reshape(np.random.random(size=y.shape), [len(y),1])
    l2 = np.reshape(getrate(rate, y, dt)/m2, [len(y),1])
    keep = np.where( prob< l2 )    
    y = y[ keep[0]]

    input = nml.SpikeArray(id=name)
    ctr = 0
    for spk in y:
        input.spikes.append(nml.Spike(id=ctr, time="{} ms".format(1e3*spk) ))

    return input


def write_combined_spikearray_nml(spike_groups, modulation_labels, filename="MF_BehMod.nml"):
    """
    Save spike trains to a NeuroML file with SpikeArrays grouped by modulation.
    """
    doc = NeuroMLDocument(id="MF_BehMod")

    for group, label in zip(spike_groups, modulation_labels):
        for i, spikes in enumerate(group):
            sa = SpikeArray(id=f"MF_{label}_{i}")
            for j, t in enumerate(spikes):
                sa.spikes.append(Spike(id=j, time=f"{t*1000:.3f} ms"))  # convert to ms
            doc.spike_arrays.append(sa)

    pynml.write_neuroml2_file(doc, filename)
    print(f"âœ… Saved spike trains to: {filename}")
```

```python
import numpy as np
import pickle
from neuroml import NeuroMLDocument
from pyneuroml import pynml

# Assumes you already defined:
# - generate_rates_from_behavior()
# - getrate()
# - create_spike_array()

# === 1. Load connectivity to get number of MFs ===
with open("../network_structures/GCLconnectivity_4.pkl", 'rb') as f:
    conn_mat = pickle.load(f, encoding='latin1')['conn_mat']
N_mf = conn_mat.shape[0]

# === 2. Define group sizes ===
n_pm = int(N_mf * 0.4)
n_nm = int(N_mf * 0.3)
n_ns = N_mf - n_pm - n_nm

# === 3. Generate behavior-modulated rates ===
minT, maxT = 0, 40  # seconds
dt = 0.0005         # seconds (0.5 ms time resolution)
modulation_labels = ["PM", "NM", "NS"]
spike_arrays_all = []

# PM: positively modulated
time_pm, rates_pm, w_pm = generate_rates_from_behavior(
    beh_file='examplebeh.txt', minT=minT, maxT=maxT, n_mf=n_pm,
    scale=50, offset=7, minrate=2, sparsity=0.1, seed=1
)
for i in range(n_pm):
    sa = create_spike_array(rates_pm[:, i], maxT=maxT, dt=dt, name=f"MF_PM_{i}")
    spike_arrays_all.append(sa)

# NM: negatively modulated
time_nm, rates_nm, w_nm = generate_rates_from_behavior(
    beh_file='examplebeh.txt', minT=minT, maxT=maxT, n_mf=n_nm,
    scale=-25, offset=30, minrate=2, sparsity=0.1, seed=2
)
for i in range(n_nm):
    sa = create_spike_array(rates_nm[:, i], maxT=maxT, dt=dt, name=f"MF_NM_{i}")
    spike_arrays_all.append(sa)

# NS: non-significantly modulated
time_ns, rates_ns, w_ns = generate_rates_from_behavior(
    beh_file='examplebeh.txt', minT=minT, maxT=maxT, n_mf=n_ns,
    scale=0, offset=18.5, minrate=2, sparsity=0.1, seed=3
)
for i in range(n_ns):
    sa = create_spike_array(rates_ns[:, i], maxT=maxT, dt=dt, name=f"MF_NS_{i}")
    spike_arrays_all.append(sa)

# === 4. Write all to NeuroML file ===
#doc = NeuroMLDocument(id="MF_BehMod")
#for sa in spike_arrays_all:
#    doc.spike_arrays.append(sa)

#pynml.write_neuroml2_file(doc, "MF_BehMod.nml")
#print("âœ… Saved behavior-modulated MF spike arrays to 'MF_BehMod.nml'")
```

```python
import matplotlib.pyplot as plt

def plot_example_rates(time_pm, rates_pm, time_nm, rates_nm, time_ns, rates_ns, n_show=5):
    """
    Plot example firing rate traces for PM, NM, and NS fibers.
    
    Parameters:
    - time_*(T,), rates_*(T x N): time vector and rate matrix for each modulation class
    - n_show: number of fibers to plot from each group
    """
    plt.figure(figsize=(15, 8))

    # Plot positively modulated (PM)
    for i in range(min(n_show, rates_pm.shape[1])):
        plt.plot(time_pm, rates_pm[:, i], label=f'PM {i}', alpha=0.8)

    # Plot negatively modulated (NM)
    for i in range(min(n_show, rates_nm.shape[1])):
        plt.plot(time_nm, rates_nm[:, i], label=f'NM {i}', alpha=0.8)

    # Plot non-significantly modulated (NS)
    for i in range(min(n_show, rates_ns.shape[1])):
        plt.plot(time_ns, rates_ns[:, i], label=f'NS {i}', alpha=0.8)

    plt.xlabel("Time (s)")
    plt.ylabel("Firing rate (Hz)")
    plt.title("Example Mossy Fiber Firing Rates")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_example_rates(time_pm, rates_pm, time_nm, rates_nm, time_ns, rates_ns)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the behavior file assuming comma-separated values
# First column: time in ms
# Remaining columns: behavioral components
data = np.genfromtxt("examplebeh.txt", delimiter=',')

# Split time and behavior
time_ms = data[:, 0]
time_s = time_ms / 1000.0  # convert to seconds
behavior_signals = data[:, 1:]

# Plot behavior components
plt.figure(figsize=(12, 6))
for i in range(behavior_signals.shape[1]):
    plt.plot(time_s, behavior_signals[:, i], label=f'Behavior dim {i+1}')

plt.xlabel("Time (s)")
plt.ylabel("Behavior signal")
plt.title("Behavioral Components Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
